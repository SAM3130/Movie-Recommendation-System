```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset into a DataFrame
df = pd.read_csv('movies_metadata.csv')

# Convert 'budget' and 'revenue' columns to numeric format
df['budget'] = pd.to_numeric(df['budget'], errors='coerce')
df['revenue'] = pd.to_numeric(df['revenue'], errors='coerce')

# Define the budget intervals
bins = [0, 1e6, 1e7, 1e8, 1e9, float('inf')]
labels = ['< 1M', '1M - 10M', '10M - 100M', '100M - 1B', '> 1B']

# Group the data by budget intervals and calculate the average revenue
df['budget_intervals'] = pd.cut(df['budget'], bins=bins, labels=labels)
avg_revenue_by_budget = df.groupby('budget_intervals')['revenue'].mean()

# Plotting the bar chart
plt.bar(avg_revenue_by_budget.index.astype(str), avg_revenue_by_budget)
plt.xlabel('Budget Intervals')
plt.ylabel('Average Revenue')
plt.title('Average Revenue by Budget Intervals')
plt.xticks(rotation=45)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset into a DataFrame
df = pd.read_csv('movies_metadata.csv')

# Convert 'budget' column to numeric format
df['budget'] = pd.to_numeric(df['budget'], errors='coerce')

# Drop rows with missing values in 'budget' and 'genres' columns
df = df.dropna(subset=['budget', 'genres'])

# Extract unique genres
unique_genres = set()
for genres_list in df['genres']:
    genres_list = eval(genres_list)  # Convert string representation to a list
    for genre in genres_list:
        unique_genres.add(genre['name'])

# Calculate the total budget for each genre
budget_by_genre = {}
for genre in unique_genres:
    budget_by_genre[genre] = df[df['genres'].str.contains(genre)]['budget'].sum()

# Sort genres based on the total budget in descending order
sorted_genres = sorted(budget_by_genre, key=budget_by_genre.get, reverse=True)

# Prepare data for plotting
genres = []
budgets = []
for genre in sorted_genres:
    genres.append(genre)
    budgets.append(budget_by_genre[genre])

# Plotting the bar chart
plt.bar(genres, budgets)
plt.xlabel('Genres')
plt.ylabel('Total Budget')
plt.title('Total Budget by Genre')
plt.xticks(rotation=45)
plt.show()
```

```python
import pandas as pd
from tabulate import tabulate

# Load the dataset into a DataFrame
df = pd.read_csv('movies_metadata.csv')

# Convert 'revenue' column to numeric format
df['revenue'] = pd.to_numeric(df['revenue'], errors='coerce')

# Drop rows with missing values in 'revenue' and 'production_companies' columns
df = df.dropna(subset=['revenue', 'production_companies'])

# Create a dictionary to store the total revenue per production company
total_revenue_by_company = {}

# Calculate the total revenue for each production company
for _, row in df.iterrows():
    production_companies = eval(row['production_companies'])  # Convert string representation to a list
    revenue = row['revenue']

    for company in production_companies:
        company_name = company['name']
        if company_name in total_revenue_by_company:
            total_revenue_by_company[company_name] += revenue
        else:
            total_revenue_by_company[company_name] = revenue

# Get the top 10 production companies by revenue
top_10_companies = pd.Series(total_revenue_by_company).nlargest(10)

# Create a list of lists for tabular representation
table_data = []
for company, revenue in top_10_companies.items():
    table_data.append([company, revenue])

# Display the top 10 production companies in tabular form
print(tabulate(table_data, headers=['Production Company', 'Total Revenue'], tablefmt='grid'))
```

```python
import pandas as pd
from tabulate import tabulate

# Load the dataset into a DataFrame
df = pd.read_csv('movies_metadata.csv')

# Drop rows with missing values in relevant columns
df = df.dropna(subset=['genres', 'original_title', 'original_language', 'popularity', 'vote_average', 'revenue', 'production_companies'])

# Convert 'vote_average', 'popularity', and 'revenue' columns to numeric format
df['vote_average'] = pd.to_numeric(df['vote_average'], errors='coerce')
df['popularity'] = pd.to_numeric(df['popularity'], errors='coerce')
df['revenue'] = pd.to_numeric(df['revenue'], errors='coerce')

# Get the top rated movie in each genre
top_movies = df.groupby('genres').apply(lambda x: x.nlargest(1, ['vote_average', 'popularity', 'revenue']))

# Select the desired columns for display
selected_columns = ['genres', 'original_title', 'original_language', 'popularity', 'vote_average', 'revenue', 'production_companies']

# Limit the top_movies to the first 5 genres
top_movies_subset = top_movies.loc[top_movies['genres'].isin(top_movies['genres'].unique()[:5])]

# Format the table
table = tabulate(top_movies_subset[selected_columns], headers=selected_columns, tablefmt='grid', showindex=False)

# Display the table
print(table)
```

```python
import pandas as pd
from tabulate import tabulate

# Load the dataset into a DataFrame
df = pd.read_csv('movies_metadata.csv')

# Filter movies based on the "Horror" genre
filtered_movies = df[df['genres'].str.contains("Horror")]

# Sort movies by vote_average and popularity in descending order
sorted_movies = filtered_movies.sort_values(['vote_average', 'popularity'], ascending=False)

# Select the desired columns for display
selected_columns = ['original_title', 'vote_average', 'popularity']

# Get the top 5 recommended horror movies
recommended_movies = sorted_movies.head(5)[selected_columns]

# Display the recommended movies in tabular form
table = tabulate(recommended_movies, headers=selected_columns, tablefmt='grid', showindex=False)

# Print the table
print("Top 5 recommended horror movies:")
print(table)
```
